```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import CategoricalNB
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import *
import matplotlib.pyplot as plt
from sklearn.feature_selection import chi2, SelectKBest
from sklearn.model_selection import train_test_split

import mlflow
import mlflow.sklearn as mlsk
```

```python
insurance = pd.read_csv("insurance.csv").drop("Unnamed: 0", axis=1)
insurance
```

```python
insurance.groupby(insurance["Accident"]).size()
```

```python
insurance.groupby(insurance["Accident"]).sum(numeric_only=True)
```

```python
print(insurance.dtypes)
```

```python
insurance = insurance.astype("category")
insurance.dtypes
```

```python
for col in insurance.columns:
    insurance.groupby(col).size().plot(kind="bar", width = 0.75)
    plt.title(f"GrÃ¡fico de Barra de: {col}")
    plt.ylabel(f"DistribuiÃ§Ã£o de {col}")
    plt.xlabel(f"Valores de {col}")
    plt.show()

    print(f"Resumo EstatÃ­stico de: {col} \n")
    print(insurance[col].describe() )

    nulos = insurance[col].isnull().sum()
    print(f" \n Nulos na Coluna: {nulos}")
```

```python
for colunas in insurance.columns:
    insurance[colunas] = insurance[colunas].cat.codes
insurance
```

```python
X = insurance.iloc[:,:-7].values
y = insurance.iloc[:,7].values
y
```

```python
X_treino, X_teste, y_treino, y_teste = train_test_split(X,y, test_size=0.3)
X_treino.shape
```

```python
modelo_rf = RandomForestClassifier(random_state=321, verbose=0)
modelo_rf.fit(X_treino, y_treino)
```

```python
previsoes_rf = modelo_rf.predict(X_teste)
previsoes_rf[:20]
```

```python
acuracia = accuracy_score(y_teste, previsoes_rf)
f1_score = f1_score(y_teste, previsoes_rf, average="macro")
recall = recall_score(y_teste, previsoes_rf, average="macro")
precision = precision_score(y_teste, previsoes_rf, average="macro")
print("AcurÃ¡cia: ", acuracia)
print("F1-Score: ", f1_score)
print("Recall: ", recall)
print("Precision: ", precision)
```

```python
confusao_rf = ConfusionMatrixDisplay.from_estimator(modelo_rf, X_teste, y_teste, cmap="summer")
```

```python
seletor = SelectKBest(chi2, k =  12)
seletor.fit(X_treino, y_treino)
X_treino = seletor.transform(X_treino)
X_teste = seletor.transform(X_teste)
```

```python
X_teste.shape
```

```python
mlflow.set_experiment("InsuranceRF")
with mlflow.start_run():
    new_model = RandomForestClassifier().fit(X_treino, y_treino)
    new_predictions = new_model.predict(X_teste)
    new_predictions[:20]
    acuracia = accuracy_score(y_teste, new_predictions)
    recall = recall_score(y_teste, new_predictions, average="macro")
    precision = precision_score(y_teste, new_predictions, average="macro")

    mlflow.log_metric("AcurÃ¡cia", acuracia)
    mlflow.log_metric("Recall", recall)
    mlflow.log_metric("Precision", precision)

    mlflow.sklearn.log_model(new_model, "RF-Insurance")

    confusao = ConfusionMatrixDisplay.from_estimator(new_model, X_teste, y_teste)
    plt.savefig("Confusao-RF.png")
    mlflow.log_artifact("Confusao-RF.png")
    print("Modelo:", mlflow.active_run().info.run_uuid)

mlflow.end_run()
print("AcurÃ¡cia: ", acuracia)
print("Recall: ", recall)
print("Precision: ", precision)
```
